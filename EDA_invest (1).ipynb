```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\ML_train\invest_db.csv", sep=';')
data.head()
```

```python
df = data.apply(lambda col: col.fillna(col.mode()[0]))
df.head(20)
```

```python
df['portf_total_amt'] = df['portf_total_amt'].str[:-2]
df
```

```python
df.drop(df[df['education_level_cd'] == 'MGR'].index, inplace=True)
```

```python
df['portf_total_amt'] = df['portf_total_amt'].astype(int)
```

```python
mn = df.groupby('education_level_cd')['portf_total_amt'].mean().round(2).reset_index()
mn.sort_values(by = 'portf_total_amt', ascending=False)
```

```python
plt.hist(df['education_level_cd'], bins=20, color='green', alpha=0.7)
plt.xlabel('Образование')
plt.ylabel('Количество людей')
plt.show()
```

```python
df['trade_order_rus_bon_cnt'] = df['trade_order_rus_bon_cnt'].str[:-2]
df['trade_order_forts_cnt'] = df['trade_order_forts_cnt'].str[:-2]
df['trade_order_rus_sec_cnt'] = df['trade_order_rus_sec_cnt'].str[:-2]
df['trade_order_cur_cnt'] = df['trade_order_cur_cnt'].str[:-2]
df['trade_order_opt_cnt'] = df['trade_order_opt_cnt'].str[:-2]
```

```python
df['trade_order_rus_bon_cnt'] = df['trade_order_rus_bon_cnt'].astype(int)
df['trade_order_forts_cnt'] = df['trade_order_forts_cnt'].astype(int)
df['trade_order_rus_sec_cnt'] = df['trade_order_rus_sec_cnt'].astype(int)
df['trade_order_cur_cnt'] = df['trade_order_cur_cnt'].astype(int)
df['trade_order_opt_cnt'] = df['trade_order_opt_cnt'].astype(int)
```

```python
df['total_sum'] = df['trade_order_rus_bon_cnt'] + df['trade_order_forts_cnt'] + df['trade_order_rus_sec_cnt'] + df['trade_order_cur_cnt'] + df['trade_order_opt_cnt']
```

```python
kolvo = df.groupby('age')['total_sum'].sum().reset_index()
kolvo.sort_values(by = 'total_sum', ascending=False)
kolvo.set_index('age', inplace=True)
```

```python
age_counts = df['age'].value_counts()
age_counts
```

```python
dff = kolvo.join(age_counts)
dff
```

```python
dff.plot(style=['g','r'])
plt.title('Распределение по количеству и сделкам')
plt.show()
```

```python
dff['q_per_person'] = dff['total_sum'] / dff['count']
dff.sort_values(by = 'q_per_person', ascending=False).round(2).head(10)
```
